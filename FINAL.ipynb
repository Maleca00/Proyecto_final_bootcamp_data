**Introducción**

La Organización Mundial de la Salud, describe la obesidad y el sobrepeso como la acumulación excesiva de grasa en determinadas zonas del cuerpo que puede ser perjudicial para la salud, el número de personas que sufre de obesidad se ha duplicado desde 1980. Algunas de las causas del sobrepeso son el aumento de la ingesta de alimentos densos en energía y ricos en grasas y la disminución de la actividad física debido a la naturaleza sedentaria de los nuevos trabajos, los nuevos nuevos medios de transporte y el aumento de la urbanización.

**Contexto comercial:** 
En la actualidad la obesidad constituye un creciente problema de salud pública. Según la OMS, están relacionados con el sobrepeso o la obesidad, a nivel global, al menos 2,8 millones de muertes. **Ingeniería de atributos**

Como se observó en el apartado de integridad de la data, las columnas no contienen valores nulos, por lo que no es requerido realizar imputación en niguna de las columnas, como tampoco se ve la necesidad de crear columnas adicionales que extraigan características o a grupaciones específicas en el dataset. Lo que si se utiliza es get dummies, dado que la mayoria de columnas son categoricas.


```python
x=datos[['Gender','Age','Height','Weight','family_history_with_overweight','FAVC','FCVC','NCP','CAEC','SMOKE','CH2O','SCC','FAF','TUE','CALC','MTRANS']]
y= datos['NObeyesdad']
```

```python
numer=x[['Age','Height','Weight','FCVC','NCP','CH2O','FAF','TUE']] #VariablesNumericas
bina=pd.get_dummies(x[['Gender','family_history_with_overweight','FAVC','CAEC','SMOKE','SCC','CALC','MTRANS']],drop_first=True) #Variables Binarias
x_new=pd.concat([numer,bina],axis=1)
x_new.head()
```

```python
y_bina=pd.get_dummies(y)
y_bina.head()
```

**Predicción**

Dado que estamos tratando con un problema de clasificación se decidió utilizar el algoritmo de predicción random forest y la regresión logística.

**RANDOM FOREST**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_new, y, test_size=0.30, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classif = RandomForestClassifier(max_depth=6, random_state=42, n_estimators=30,criterion='gini')
classif.fit(X_train, y_train)
```

```python
y_pred= classif.predict(X_test)
```

**Metricas Random Forest**

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred=y_pred, y_true= y_test))
```

```python
sorted_idx = classif.feature_importances_.argsort()
plt.barh(X_train.columns[sorted_idx], classif.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
```

**REGRESIÓN LOGÍSTICA**

```python
from sklearn.linear_model import LogisticRegression
classif2 = LogisticRegression(random_state=42,max_iter=500,solver='newton-cg').fit(X_train, y_train)
```

```python
y_pred2=classif2.predict(X_test)
```

**Metricas Regresión Logística**

```python
print(classification_report(y_pred=y_pred2, y_true= y_test))
```

**¿QUÉ DICEN LOS MODELOS?**

Para el algoritmo Random Forest se evidencia que la exactitud presentada es del 86% y para todos los tipos de obesidad la precisión es igual o superior al 75%, en donde la menor probabilidad se obtuvo en la variable Normal_Weight, mientras que la mayor se obtuvo en obesity_Type_III, lo que significa que es más sensible a la obesidad tipo 3 pero le cuesta más clasificar el peso normal y que en general más del 70% de predicciones son acertadas; adicionalmente, para todas las categorías el algoritmo tiene un F1-score entre el 76% y el 99%, lo que quiere decir que existe un buen balance entre la precisión y el recall. Esto es importante dado que el F1 score es especialmente útil en situaciones donde se desea tener un balance adecuado entre la capacidad del modelo para identificar correctamente las instancias positivas y la capacidad para evitar falsos positivos. 
En general, estos resultados indican que el modelo tiene un buen rendimiento en la mayoría de las clases, con altos valores de precisión, recall y F1-score. El valor de precisión global (accuracy) también es bastante alto, lo que sugiere que el modelo es efectivo en la clasificación de las diferentes clases. 

Adicionalmente, en este algoritmo se puede verificar la importancia de las características utilizadas en el modelo, inicialmente es importante recordar que el IMC (medida que nos da la clasificación del tipo de obesidad) es una medida simplificada que no tiene en cuenta todos los factores que influyen en la obesidad. Además, la relación entre estas variables y el IMC puede ser influenciada por otros factores individuales, como la genética. La obesidad es una condición multifactorial, y estas variables son solo algunas de las que pueden influir en el IMC. En este caso tiene sentido que lo más importante sea el peso dado que el IMC evalúa la relación entre el peso y la altura de una persona como también tiene sentido que (como se mencionó antes), variables como el medio de transporte o el consumo de cigarrillo tuvo poca importancia por la naturaleza de los datos que se tenían en esas columnas

Por otra parte, se tiene la regresión logística que tiene un porcentaje exactitud de 84% y de presición igual o mayor que 71%, enlal cual el algoritmo es menos preciso para predecir la variable Overweight_Level_I y su más preciso para obesity_Type_III, igual que el random forest, a diferencia del anterior algoritmo, se evidencia que el f1-score está entre el 70% y el 99%, lo que significa que para algunas de las variables a clasificar el balance entre la precisión y el recall es menor que en el modelo anerior. En resumen, estos resultados indican cómo el modelo se desempeña en términos de precisión, exhaustividad y equilibrio entre ambas métricas para cada clase. Los valores de precisión y recall varían según la clase, lo que sugiere que el modelo es mejor para algunas clases que para otras.

Se comparan estos dos tipos de algoritmos de clasificación, con el F1 score dado que es la que nos indica con mayor fidelidad la armonía en las clasificaciones, obteniendo que el Random forest tiene en general un porcentaje mayor que la Regresión Logística, por lo tanto es mejor clasificando la data de estudio en cada clasificación de la obesidad, adicionalmente el algoritmo Random Forest está mejor optimizado para datos como los utilizados, que son no lineales o desequilibrados, también es menos sensible a características irrelevantes, como en este fue el medio de transporte o el consumo de cigarrillo.
